```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
def fill_missing_by_group_mean(df, group_col, target_cols):
    for col in target_cols:
        df[col] = df[col].fillna(df.groupby(group_col)[col].transform('mean'))
    return df


# Fill missing values in population columns
df4 = fill_missing_by_group_mean(df, 'Continent', df.select_dtypes(include='number').columns)
```

```python
df4.isnull().sum()
```

```python
df4.drop(columns='2022 Population', axis=1)
```

```python
df4['Continent']=df4['Continent'].str.strip()
```

```python
df4['Country']=df4['Country'].replace('[^a-zA-Z]', '')
```

```python
df4['Country']=df4['Country'].apply(lambda x: x.upper())
```

```python
df4[['Country1', 'Country2']] = df4['Country'].str.split(' ', n=1,expand=True)
```
