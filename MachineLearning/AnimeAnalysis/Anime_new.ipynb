# Data Encoding

```python
Y = anime_cleaned['score']
X = anime_cleaned.drop(['score','username','user_id','title','genre'], axis=1)
X_encoded = pd.get_dummies(X, columns=['gender', 'type', 'source'])
print(X_encoded.head())
```

```python

from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

seed = 7
# Split the dataset (both input & output) into training/testing datasets
X_train, X_test, Y_train, Y_test= train_test_split(X_encoded,Y, test_size=0.2, random_state=seed)
```

# Data Modeling

## CART

```python
#CART
# Train the model using the training sub-dataset
model = DecisionTreeRegressor(random_state=seed)
dtr=model.fit(X_train,Y_train)

fig = plt.figure(figsize=(70,50))
tree.plot_tree(dtr,filled = True, fontsize=15)
plt.show()


R_squared = model.score(X_test, Y_test)
print('R-Squared = ', R_squared)


num_folds = 10
seed = 7
# Split the whole data set into folds
kfold= KFold(n_splits=num_folds, random_state=seed, shuffle=True)
scoring = 'neg_mean_squared_error'

# Train the model and run K-foLd cross-validation to validate/evaluate the model
results = cross_val_score(model, X_encoded, Y, cv=kfold, scoring=scoring)
print("Average of all results from the K-fold Cross Validation, using negative mean squared error:",results.mean())
```

## KNN

```python
#KNN
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import classification_report

model = KNeighborsRegressor(n_neighbors=3)
model.fit(X_train, Y_train)
y_pred = model.predict(X_test)
r_squared = model.score(X_test,Y_test)

print("R-squared:", r_squared)

# evaluate the algorithm
# specify the number of time of repeated splitting, in this case 10 folds
n_splits = 10
seed = 7
kfold = KFold(n_splits=n_splits, random_state=seed, shuffle=True)
scoring = 'r2'
results = cross_val_score(model, X_encoded, Y, cv=kfold, scoring=scoring)
print("R-squared (Cross-Validated): %.3f (%.3f)" % (results.mean(), results.std()))
```

## Linear Regression

```python
# Build the Linear Regression model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
# Train the model using the training sub-dataset
model.fit(X_train, Y_train)
# Print intercept and coefficients
print ("Intercept:", model.intercept_)
print ("Coefficients:", model.coef_)

R_squared = model.score(X_test, Y_test)
print("R-squared: ", R_squared)
```
